```python
## Plot file name
save_fn = dataset+"_stoc_reward_"+str(stoc_reward)+".pdf"

####### Data collection:
data = []
for i, n in enumerate(num):
    data.append(read_csv("../logs/"+logs_path+"/se_num_data_"+str(n)+\
                            "_random_state_"+str(random_state[0])+\
                            "_stoc_reward_"+str(stoc_reward)+".csv"))
estimators = data[0][data[0].columns[0]].tolist()

print(estimators)
avg_data = []
std_data = []
for i, n in enumerate(num):
    temp = []
    for j, rs in enumerate(random_state):
        d = read_csv("../logs/"+logs_path+"/se_num_data_"+str(n)+\
                                "_random_state_"+str(rs)+\
                                "_stoc_reward_"+str(stoc_reward)+".csv")
        temp.append(d["mean"])
    temp=np.array(temp)
    temp1 = []
    temp2 = []
    for j, _ in enumerate(estimators):
        temp1.append(np.mean(temp[np.argpartition(temp[:,j], n_seeds)[:n_seeds],j]))
        temp2.append(np.std(temp[np.argpartition(temp[:,j], n_seeds)[:n_seeds],j]))
        
    avg_data.append(temp1)
    std_data.append(temp2)
    
if num_last:
    temp = []
    for j, rs in enumerate(random_state_last):
        d = read_csv("../logs/"+logs_path+"/se_num_data_"+str(num_last)+\
                                "_random_state_"+str(rs)+\
                                "_stoc_reward_"+str(stoc_reward)+".csv")
        temp.append(d["mean"])
    temp=np.array(temp)
    avg_data.append(np.mean(temp,axis=0))
    std_data.append(np.std(temp,axis=0))
    
avg_data = np.array(avg_data)
std_data = np.array(std_data)
####### Set colors:
colors = ['black', 'firebrick', 'orangered', 'orange', 'gold', 
          'yellowgreen', 'turquoise', 'deepskyblue', 'green', 'navy',
          'indigo', 'magenta', 'slategray']
bands = ['lightgrey', 'mistyrose', 'peachpuff', 'moccasin', 'khaki',
         'palegreen', 'lightcyan', 'powderblue', 'mediumspringgreen', 'lavender',
        'thistle', 'lavenderblush', 'gainsboro']
markers = [".","v","1","8","s",
           "P","^","X","d","<",
           "3","*",">"]
```

```python
# plt.figure()
fig, ax = plt.subplots()
for idx, est in enumerate(estimators):
#     print(num)
    avg = []
    stds = []
    for i, n in enumerate(num):
        avg.append(avg_data[i][idx])
        stds.append(std_data[i][idx])
    if num_last:
        avg.append(avg_data[-1][idx])
        stds.append(std_data[-1][idx])
        
    avg = np.array(avg)
    stds = np.array(stds)
    
    if num_last:
        if est=='DR-IC (tau=0)':
            plt.plot(num_, avg, label=r'dr-ic ($\tau=0$)' , color=colors[idx], marker=markers[idx])
        else:
            plt.plot(num_, avg, label=est , color=colors[idx], marker=markers[idx])
        plt.fill_between(num_, avg - stds, avg + stds , color=bands[idx] ) 
    else:
        if est=='DR-IC (tau=0)':
            plt.plot(num, avg, label=r'dr-ic ($\tau=0$)' , color=colors[idx], marker=markers[idx])
        else:
            plt.plot(num, avg, label=est , color=colors[idx], marker=markers[idx])
        plt.fill_between(num, avg - 0.5*stds, avg + 0.5*stds , color=bands[idx] ) 

plt.xlabel("n",fontsize=15)
plt.ylabel("MSE",fontsize=15)
plt.yscale('log')
# plt.xscale('log')
# plt.ylim(3e-5,2e-2)
# lgd = plt.legend(loc = 'lower left', bbox_to_anchor=(1.2, 1.))
# lgd = plt.legend(loc = 'lower left', ncol=13, bbox_to_anchor=(0., -0.5))

if stoc_reward:
    plt.title(dataset+", stochastic reward",fontsize=15)
else:
    plt.title(dataset+", deterministic reward",fontsize=15)
    
# ymin, ymax = plt.ylim()
# # ax.set_yticks([round(ymin,3), round((ymax-ymin)/4,3), round(ymax,3)])
# ax.set_yticks([round(ymax,3)])
# s=[]
# # s.append(np.format_float_scientific(round(ymin,3), unique=False, precision=1, exp_digits=1))
# # s.append(np.format_float_scientific(round((ymax-ymin)/4,3), unique=False, precision=1, exp_digits=1))
# s.append(np.format_float_scientific(round(ymax,3), unique=False, precision=1, exp_digits=1))
# plt.setp(ax.set_yticklabels(s), rotation=0)

plt.grid()
ax.yaxis.set_label_coords(-0.03,0.6)
    
plt.savefig("../logs/"+logs_path+"/"+save_fn, dpi = 1000, transparent=True,bbox_inches='tight') # final use

# plt.savefig("../logs/"+logs_path+"/"+save_fn, bbox_extra_artists=(lgd,), bbox_inches='tight') # Use this when you need a pdf.
# plt.savefig("../logs/glass/glass_test.pdf", dpi = 1000, bbox_extra_artists=(lgd,), bbox_inches='tight') # Use this when you need a pdf.
# plt.savefig("../logs/glass/glass_test.pdf", bbox_extra_artists=(lgd,), bbox_inches='tight') # Use this when you need a pdf.
# plt.show()
```
